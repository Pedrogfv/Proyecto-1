Comenzamos con los primeros parametros:

Nuestro precio es mayor o menor que la competencia? 

```python
# Unir los DataFrames basados en la columna 'Referencia'
df_comparativo = pd.merge(df_2[['Referencia', 'Nombre', 'Nuestro Precio']], 
                          df_3[['Referencia', 'Precio día 2']], 
                          on='Referencia')

# Añadir una columna de comparación
df_comparativo['Comparativa'] = df_comparativo['Nuestro Precio'] - df_comparativo['Precio día 2']

# Definir si el precio es mayor o menor
df_comparativo['Resultado'] = df_comparativo['Comparativa'].apply(lambda x: 'Mayor' if x > 0 else 'Menor')

# Mostrar el resultado
print(df_comparativo[['Referencia', 'Nombre', 'Nuestro Precio', 'Precio día 2', 'Resultado']])
```

El precio de la competencia cambió de un día para otro?

```python
# Filtrar las columnas necesarias de 'Referencia', 'Nombre', 'Ventas Enero', y 'Ventas Febrero'
df_precios = df_3[['Referencia', 'Nombre', 'Precio día 1', 'Precio día 2']]

# Crear una columna que compare si el precio cambió entre enero y febrero
df_precios['Resultado'] = df_precios.apply(
    lambda row: 'Sí cambió el precio' if row['Precio día 1'] != row['Precio día 2'] else 'No cambió el precio', axis=1
)

# Mostrar el resultado con las columnas de 'Referencia', 'Nombre', y el resultado de la comparación
print(df_precios[['Referencia', 'Nombre', 'Resultado']])
```

Multiplicamos por 0.99 nuestros productos que tienen el precio mayor y que la competencia cambió su precio

```python
# Datos de tu empresa: seleccionamos la referencia, nombre, y nuestro precio
df_precios_empresa = df_2[['Referencia', 'Nombre', 'Nuestro Precio']]

# Datos de la competencia: seleccionamos la referencia, precio día 1 y precio día 2
df_precios_competencia = df_3[['Referencia', 'Precio día 1', 'Precio día 2']]

# Unir los DataFrames por 'Referencia'
df_comparativa = pd.merge(df_precios_empresa, df_precios_competencia, on='Referencia')

# Crear una columna para comparar si nuestro precio es mayor que el precio del día 2 de la competencia
df_comparativa['Precio Mayor'] = df_comparativa['Nuestro Precio'] > df_comparativa['Precio día 2']

# Crear una columna para verificar si el precio cambió en la competencia (día 1 a día 2)
df_comparativa['Cambio Precio Competencia'] = df_comparativa['Precio día 1'] != df_comparativa['Precio día 2']

# Aplicar la condición de multiplicar por 0.99 si ambas condiciones son verdaderas
df_comparativa['Nuevo Precio'] = df_comparativa.apply(
    lambda row: row['Nuestro Precio'] * 0.99 if row['Precio Mayor'] and row['Cambio Precio Competencia'] else row['Nuestro Precio'],
    axis=1
)

# Filtrar únicamente las filas donde el "Nuevo Precio" es diferente del "Nuestro Precio" (los que se multiplicaron por 0.99)
df_multiplicados = df_comparativa[df_comparativa['Nuevo Precio'] < df_comparativa['Nuestro Precio']]

# Mostrar solo los productos cuyo precio fue multiplicado por 0.99
print(df_multiplicados[['Referencia', 'Nombre', 'Nuestro Precio', 'Nuevo Precio']])
```

SAcar utilidad de los productos con nuevo precio

```python
# Ahora hacemos el merge con el DataFrame 'df' que contiene los costos 
df_merged = pd.merge(df_multiplicados, df[['Referencia', 'Costo de Producto']], on='Referencia')

# Calcular el porcentaje de utilidad con la fórmula (Nuevo Precio - Costo) / Costo * 100
df_merged['Porcentaje de Utilidad'] = ((df_merged['Nuevo Precio'] - df_merged['Costo de Producto']) / df_merged['Costo de Producto']) * 100

# Mostrar el resultado con las columnas 'Referencia', 'Nombre', 'Nuevo Precio', 'Costo de Producto', y 'Porcentaje de Utilidad'
print(df_merged[['Referencia', 'Nombre', 'Nuevo Precio', 'Costo de Producto', 'Porcentaje de Utilidad']])
```

Buscamos el inventario de los productos seleccionados

```python
# Creamos un Dataframe combinando 'df' y 'df_multiplicados'
df_inventario = pd.merge(df_multiplicados, df[['Referencia', 'Inventario Actual']], on = 'Referencia')

print(df_inventario[['Referencia', 'Nombre', 'Inventario Actual']])
```

Quién vende más producto? Online tiene que vender más del 75% de lo que vende Sucursal

```python
df_os = pd.merge(df_multiplicados, df_2[['Referencia', 'Vendidos día 2 Online', 'Vendidos día 2 Sucursal']], on = 'Referencia')

df_os['Quién Vende más?'] = df_os.apply(
    lambda row: 'Sucursal' if row['Vendidos día 2 Sucursal']*1.75 > row ['Vendidos día 2 Online'] else 'Online', axis=1 
    )

print(df_os[['Referencia', 'Nombre', 'Quién Vende más?']])
```


Continuamos con los segundos parámetros:

- Cambió el precio de la competencia de un día para otro?
- Nuestro producto tiene menor o mayor precio que la competencia?
- Sí el precio de la competencia cambió, cambió más del 10%?

```python
# Unir los DataFrames por 'Referencia'
df_comparacion = pd.merge(df_2[['Referencia', 'Nombre', 'Nuestro Precio']], df_3[['Referencia', 'Precio día 1', 'Precio día 2']], on = 'Referencia')

# Crear una columna para comparar si nuestro precio es menor que el precio del día 2 de la competencia
df_comparacion['Nuestro Precio Menor'] = df_comparacion['Nuestro Precio'] < df_comparacion['Precio día 2']

# Crear una columna para verificar si el precio cambió en la competencia (día 1 a día 2)
df_comparacion['Cambio Precio Competencia'] = df_comparacion['Precio día 1'] != df_comparacion['Precio día 2']

# Filtrar únicamente los productos donde:
# - "Nuestro Precio" es menor
# - La competencia cambió sus precios del día 1 al día 2
df_filtrados = df_comparacion[(df_comparacion['Nuestro Precio Menor']) & (df_comparacion['Cambio Precio Competencia'])]

# Calcular el porcentaje de crecimiento del precio de la competencia
df_filtrados['Porcentaje Crecimiento'] = ((df_filtrados['Precio día 2'] - df_filtrados['Precio día 1']) / df_filtrados['Precio día 1']) * 100

# Mostrar el resultado con las columnas 'Referencia', 'Nombre', 'Nuestro Precio', 'Precio día 1', 'Precio día 2', y 'Porcentaje Crecimiento'
df_filtrados[['Referencia', 'Nombre', 'Precio día 1', 'Precio día 2', 'Porcentaje Crecimiento', 'Nuestro Precio']]
```
